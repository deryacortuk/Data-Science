```python
# Data Standardization
```

```python
import numpy as np
import pandas as pd
```

```python
d1 = np.array([2,3,1,7])
d2 = np.array([7,8,9,5])
d3 = np.array([13,11,19,17])
```

```python
df = pd.DataFrame({
    "D1":d1, "D2":d2, "D3":d3
})

df = df.astype(float)
```

```python
df
```

```python
from sklearn import preprocessing
```

```python
preprocessing.scale(df)  # Standardization
```

```python
# Normalization (0-1)
```

```python
preprocessing.normalize(df)
```

```python
scaler = preprocessing.MinMaxScaler(feature_range=(10,30))
```

```python
scaler.fit_transform(df)
```
